```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
iris = sns.load_dataset('iris')
iris.head()
```

```
iris.groupby(['species']).mean() # species에 따라 각 column의 평균값이 출력됨
```

```
iris.groupby(['species']).describe()
```

```
h = iris.hist()
```

Pairplot
- 범주형인 species 기준으로 분류해서 수치값인 나머지 Column들을 그래프에 그린다
- 대각선방향의 그래프경우 x, y축의 항목이 동일하므로 자기자신에 대한 그래프를 그린다


```
sns.pairplot(iris) 
```

```
sns.pairplot(iris,  hue = 'species') 
```

```
sns.pairplot(iris,  hue = 'species', kind='reg') 
```

```
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6);
```

```
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=10); # 등고선의 갯수가 늘어남
```

```
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, colors='C3')
```

ridgepltot

```
g = sns.FacetGrid(iris, row='species', height=1, aspect=3) # subplots와 같이 그래프 갯수를 설정값만큼 만들수 있다
g.map(sns.kdeplot, 'sepal_length', shade = True) # shade는 그래프안에 색을 칠한다
```
